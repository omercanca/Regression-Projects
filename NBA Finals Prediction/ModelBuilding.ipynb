```python
"""
Created on 17 June 2022

@author: Omer Canca
"""

import pandas as pd
import numpy as np
import statsmodels.api as sm

# Reading in data
df = pd.read_csv('nbadata.csv')

# Defining response and predictor variables
pred = ['Home','MP','FG','FGA','FGP','TP','TPA','TPP','FT','FTA','FTP','ORB','DRB','TRB','AST','STL','BLK','TOV', 'PTS']
y = df["Win"]

# Model Builder Function for OLS model with selected variables as forward selection continues
def modelbuilder():
    x = df[pred]
    summ = sm.GLM(y, x, family=sm.families.Binomial()).fit()
    print(summ.summary())
modelbuilder()

# Forward Selection: Removing least significant variable, one-by-one, until all p-values are at or less than 0.05
pred.remove("MP")
modelbuilder()

pred.remove("FTA")
modelbuilder()

pred.remove("FGP")
modelbuilder()

pred.remove("Home")
modelbuilder()

pred.remove("TPP")
modelbuilder()

pred.remove("AST")
modelbuilder()

pred.remove("FG")
modelbuilder()

pred.remove("BLK")
modelbuilder()

pred.remove("FTP")
modelbuilder()

```

```python
# Final Model Creation
z = df[pred]
finalmodel = sm.GLM(y, z, family=sm.families.Binomial()).fit()
print(finalmodel.summary())
```
